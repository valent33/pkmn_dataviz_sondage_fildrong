```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
link = 'https://bulbapedia.bulbagarden.net/wiki/Bulbasaur_(Pok%C3%A9mon)'
Gen = 1
Legendary = False

LegList = ["Mew","Celebi","Jirachi","Deoxys","Phione","Manaphy","Darkrai","Shaymin","Arceus","Victini","Keldeo","Meloetta","Genesect","Diancie","Hoopa","Volcanion","Magearna","Marshadow","Zeraora","Meltan","Melmetal","Zarude",
"Mewtwo","Lugia","Ho-Oh","Kyogre","Groudon","Rayquaza","Dialga","Palkia","Giratina","Reshiram","Zekrom","Kyurem","Xerneas","Yveltal","Zygarde","Cosmog","Cosmoem","Solgaleo","Lunala","Necrozma","Zacian","Zamazenta","Eternatus","Calyrex",
"Articuno","Zapdos","Moltres","Raikou","Entei","Suicune","Regirock","Regice","Registeel","Latias","Latios","Uxie","Mesprit","Azelf","Heatran","Regigigas","Cresselia","Cobalion","Terrakion","Virizion","Tornadus","Thundurus","Landorus","Tapu Koko","Tapu Lele","Tapu Bulu","Tapu Fini","Kubfu","Urshifu","Regieleki","Regidrago","Glastrier","Spectrier","Enamorus"]

df = pd.DataFrame(columns=["No","Name","Type1","Type2","Total","HP","Attack","Defense","Sp. Atk","Sp. Def","Speed","Generation","Legendary"])

for i in range(1,906):
    # print(link)
    r = requests.get(link)
    # print(r)
    soup = BeautifulSoup(r.content, 'html.parser')
    # print(soup.prettify())

    if i == 771:
        next = "Type:_Null"
    elif 783<i<788 or i==865 or i==438 or i==121:
        next = soup.find_all('span', style="color:#000;")[3].string.split(' )')[0].split(': ')[-1]
    else:
        next = soup.find_all('span', style="color:#000;")[3].string.split(' ')[1]
    link = f'https://bulbapedia.bulbagarden.net/wiki/{next}_(Pok%C3%A9mon)'

    types = soup.find_all('span', style="color:#FFF;")
    Type1 = types[0].string
    Type2 = types[1].string

    other_info = soup.find_all('table', attrs={'class':"roundy", 'style':"background:#FFF;", 'width': '100%'})
    
    Color = other_info[-2].tbody.tr.td.contents[-1].replace('\n', '').replace(' ', '')
    if Color == '':
        Color = other_info[-2].tbody.tr.td.contents[1].replace('\n', '').replace(' ', '')
    # Color = other_info[-2].tbody.tr.td.contents[-1].replace('\n', '').replace(' ', '') # Color
    if other_info[3].tbody.tr.contents[-1].string is None:
        Weight = other_info[5].tbody.tr.contents[-1].string.replace(' kg\n', '') # Weight
        Height = other_info[4].tbody.tr.contents[-1].string.replace(' m\n', '') # Height
    else:
        Weight = other_info[4].tbody.tr.contents[-1].string.replace(' kg\n', '') # Weight
        Height = other_info[3].tbody.tr.contents[-1].string.replace(' m\n', '') # Height


    stats = soup.find_all('div', style="float:right")
    Hp = int(stats[0].contents[0])
    Attack = int(stats[1].contents[0])
    Defense = int(stats[2].contents[0])
    SpAtk = int(stats[3].contents[0])
    SpDef = int(stats[4].contents[0])
    Speed = int(stats[5].contents[0])
    Total = int(stats[6].contents[0])
    
    if i == 772 or 784<i<789 or i==866 or i==439 or i==122:
        Name = soup.title.contents[0].split(' (')[0]
    else:
        Name = soup.title.contents[0].split(' ')[0]

    if i == 151:
        Gen = 2
    elif i == 251:
        Gen = 3
    elif i == 386:
        Gen = 4
    elif i == 493:
        Gen = 5
    elif i == 649:
        Gen = 6
    elif i == 721:
        Gen = 7
    elif i == 810:
        Gen = 8

    if Name in LegList:
        Legendary = True
    else: Legendary = False

    temp = pd.Series([i, Name, Type1, Type2, Total, Hp, Attack, Defense, SpAtk, SpDef, Speed, Gen, Legendary, Color, Weight, Height],
               index=["No","Name","Type1","Type2","Total","HP","Attack","Defense","Sp. Atk","Sp. Def","Speed","Generation","Legendary", "Color", "Weight", "Height"])
    df = df.append(temp, ignore_index=True)

print(df.head())
```

```python
df = df.astype({'Total': int,
                'HP': int,
                'Attack': int,
                'Defense': int,
                'Sp. Atk': int,
                'Sp. Def': int,
                'Speed': int,
                'Generation': int,
                'Legendary': bool,
                'Height': float,
                'Weight': float})
```
